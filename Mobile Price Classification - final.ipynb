```python
X = train.drop('price_range',axis=1)
y = train['price_range']
```

```python
from sklearn.model_selection import train_test_split

#Splitting data into train and test set 80/20
#stratify - evenly distribute values, select random
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=100, test_size=0.2, stratify=y)
print(X_train.shape, X_test.shape)
#X.head()
```

# Model Build

```python
#A decision tree is an easily interpretable algorithm and is a great starting point for this use case. 
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
#training
dt.fit(X_train, y_train)
#testing
y_pred_dt = dt.predict(X_test)
```

```python
#Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
```

```python
# XGBoost Classifier
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix
#To see how the model misclassified the results, lets use Matplotlib and pandas to visualize our confusion matrix
from sklearn.metrics import accuracy_score
```

```python
#confusion matrix Decision Tree
#type1 error = 10 falsely classified as non fraud transaction, actually was fraud
#type2 error = 
dt_acc = accuracy_score(y_pred_dt, y_test)
print("Accuracy score of Decision Tree Classifier Algorithm is: ", dt_acc)
print(confusion_matrix(y_pred_dt,y_test))
```

```python
rf_acc = accuracy_score(y_pred_rf, y_test)
print("Accuracy score of Random Forest Classifier Algorithm is: ", rf_acc)
print(confusion_matrix(y_pred_rf,y_test))
```

```python
xgb_acc = accuracy_score(y_pred_xgb, y_test)
print("Accuracy score of XGBoost Classifier Algorithm is: ", xgb_acc)
print(confusion_matrix(y_pred_xgb,y_test))
```

```python
plt.bar(x=['dtc','rfc','xgb'],height=[dt_acc,rf_acc,xgb_acc])
plt.xlabel("Algorithms")
plt.ylabel("Accuracy Score")
plt.show()
```
